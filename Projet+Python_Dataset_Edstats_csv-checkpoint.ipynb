```python
fichier1 = pd.read_csv('EdStatsData.csv') 

# Sur le graphique ci dessus on voit que la colonne 'Unnamed: 69' ne contient aucune données. On va donc la supprimer.

fichier1 = fichier1.drop(columns=["Unnamed: 69"])

# Aussi les années allant de 1970 à 2000 ne nous intéressent pas car à cette époque les individus n'avaient pas accès à Internet.
# Nous allons donc garder les années allant de 2000 à 2040 car au-dessus d'ici 2040 la situation aura certainement évolué et les
# prévisions pour les années futures (jusqu'à 2100) auront changé.

# On supprime toutes les années (colonnes) allant de 1970 jusqu'à 2015

fichier1 = fichier1.drop(list(map(lambda x : str(x), list(range(1970,2000)))), axis=1)
fichier1 = fichier1.drop(list(map(lambda x : str(x), list(range(2045,2100,5)))), axis=1)
fichier1 = fichier1.drop(['2100'], axis=1)
fichier1 = fichier1.drop(['Country Code', 'Indicator Code'], axis=1)

fichier1
```

```python
# Nous allons importer le fichier 'EdStatsCountry.csv'. En effet ce fichier contient des informations par rapport au niveau de
# richesse des pays. On va pouvoir ainsi affiner notre base de données en ne retenant que des pays ayant un revenu suffisamment
# élevé pour être considéré comme stable sur le plan économique et politique.

fichier2 = pd.read_csv('EdStatsCountry.csv')

fichier2['Income Group'].unique()
```

```python
Selection_Pays_nonOECD = np.array(fichier2.loc[fichier2['Income Group'] == 'High income: nonOECD','Table Name'])
Selection_Pays_Low_Income = np.array(fichier2.loc[fichier2['Income Group'] == 'Low income','Table Name'])
Selection_Pays_Upper_middle_income = np.array(fichier2.loc[fichier2['Income Group'] == 'Upper middle income','Table Name'])
Selection_Pays_Lower_middle_income = np.array(fichier2.loc[fichier2['Income Group'] == 'Lower middle income','Table Name'])
Selection_Pays_OECD = np.array(fichier2.loc[fichier2['Income Group'] == 'High income: OECD','Table Name'])


print("Pays riches non membres de l'OCDE:\n")
print(Selection_Pays_nonOECD)

print("\n\n")

print("Pays à faible revenu:\n")
print(Selection_Pays_Low_Income)

print("\n\n")

print("Pays au revenu moyen/élevé:\n")
print(Selection_Pays_Upper_middle_income)

print("\n\n")

print("Pays au revenu faible/élevé revenu:\n")
print(Selection_Pays_Lower_middle_income)

print("\n\n")

print("Pays riches membres de l'OCDE:\n")
print(Selection_Pays_OECD)

print("\n\n")
```

```python
# Nous allons maintenant étudier l'évolution future de l'indicateur 'Upper Secondary' pour les pays suivants:
# China, Switzerland, Brazil, United Kingdom, Netherlands    afin de suivre l'évolution de ces pays potentiels
# concernant l'indicateur Upper_Secondary


Upper_Secondary = fichier1.loc[(fichier1["Indicator Name"] == "Wittgenstein Projection: Percentage of the population age 15-19 by highest level of educational attainment. Upper Secondary. Total")
                              |(fichier1["Indicator Name"] == "Wittgenstein Projection: Percentage of the population age 20-24 by highest level of educational attainment. Upper Secondary. Total"),:]



for i in range(2000,2017):
    Upper_Secondary = Upper_Secondary.drop([str(i)], axis=1)
    
Upper_Secondary = Upper_Secondary.drop(["2017"], axis=1)
Upper_Secondary = Upper_Secondary.dropna()


Country_Name = []

Annee = []

Croissance = []

for i in range(0,138,2):
    
    Num_2020 = (Upper_Secondary.loc[Upper_Secondary.index[i],"2020"] + Upper_Secondary.loc[Upper_Secondary.index[i+1],"2020"])/2
    Num_2025 = (Upper_Secondary.loc[Upper_Secondary.index[i],"2025"] + Upper_Secondary.loc[Upper_Secondary.index[i+1],"2025"])/2
    Num_2030 = (Upper_Secondary.loc[Upper_Secondary.index[i],"2030"] + Upper_Secondary.loc[Upper_Secondary.index[i+1],"2030"])/2
    Num_2035 = (Upper_Secondary.loc[Upper_Secondary.index[i],"2035"] + Upper_Secondary.loc[Upper_Secondary.index[i+1],"2035"])/2
    Num_2040 = (Upper_Secondary.loc[Upper_Secondary.index[i],"2040"] + Upper_Secondary.loc[Upper_Secondary.index[i+1],"2040"])/2

    Taux_Croissance_2025 = ((Num_2025 - Num_2020)/(Num_2020))*100
    Taux_Croissance_2030 = ((Num_2030 - Num_2025)/(Num_2025))*100
    Taux_Croissance_2035 = ((Num_2035 - Num_2030)/(Num_2030))*100
    Taux_Croissance_2040 = ((Num_2040 - Num_2035)/(Num_2040))*100
    
    Croissance.append(Taux_Croissance_2025)
    Croissance.append(Taux_Croissance_2030)
    Croissance.append(Taux_Croissance_2035)
    Croissance.append(Taux_Croissance_2040)
    
    Country_Name.append(Upper_Secondary.loc[Upper_Secondary.index[i],"Country Name"])
    Country_Name.append(Upper_Secondary.loc[Upper_Secondary.index[i],"Country Name"])
    Country_Name.append(Upper_Secondary.loc[Upper_Secondary.index[i],"Country Name"])
    Country_Name.append(Upper_Secondary.loc[Upper_Secondary.index[i],"Country Name"])
    
    Annee.append('2025')
    Annee.append('2030')
    Annee.append('2035')
    Annee.append('2040')




pd.options.display.max_rows = None

d = {'Country Name': Country_Name, 'Upper Secondary 15-24 ans (Croissance en %)': Croissance, 'Année': Annee}

Upper_Secondary_15_24_df = pd.DataFrame(d)
Upper_Secondary_15_24_df = Upper_Secondary_15_24_df.dropna()
Upper_Secondary_15_24_df = Upper_Secondary_15_24_df.loc[(Upper_Secondary_15_24_df["Country Name"] == "China")
                                                        |(Upper_Secondary_15_24_df["Country Name"] == "Switzerland")
                                                        |(Upper_Secondary_15_24_df["Country Name"] == "Brazil")
                                                        |(Upper_Secondary_15_24_df["Country Name"] == "United Kingdom")
                                                        |(Upper_Secondary_15_24_df["Country Name"] == "Netherlands"),:]

Upper_Secondary_15_24_df

sns.set_theme(style="whitegrid")
fig = plt.figure(figsize=(20,8))
sns.lineplot(x="Année", y="Upper Secondary 15-24 ans (Croissance en %)", data=Upper_Secondary_15_24_df, hue = "Country Name")
```

### Metriques de régression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import *
```

```python
y = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])

print('MAE:', mean_absolute_error(y,y_pred))
print('MSE:', mean_squared_error(y,y_pred))
print('RMSE:',np.sqrt(mean_squared_error(y,y_pred)))
```

```python
# Coefficient de détermination
print('R2:', r2_score(y,y_pred))
```
